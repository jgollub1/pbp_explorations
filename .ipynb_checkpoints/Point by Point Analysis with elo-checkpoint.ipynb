```python
# NOTE: place all these functions in a .py script and import them to save space in this ipynb
import pandas as pd
import numpy as np
from helper_functions import *

df = pd.read_csv('elo_pbp.csv')
del df['Unnamed: 0']

df['p0_name'] = [df['l_name'][i] if df['winner'][i] else df['w_name'][i] for i in xrange(len(df))]
df['p1_name'] = [df['w_name'][i] if df['winner'][i] else df['l_name'][i] for i in xrange(len(df))]
df['p0_elo'] = [df['l_elo'][i] if df['winner'][i] else df['w_elo'][i] for i in xrange(len(df))]
df['p1_elo'] = [df['w_elo'][i] if df['winner'][i] else df['l_elo'][i] for i in xrange(len(df))]
df['elo_diff'] = [df['p0_elo'][i] - df['p1_elo'][i] for i in xrange(len(df))]
df = df[['tourney_id','tourney_name','tourney_date','p0_name','p1_name','p0_elo','p1_elo','elo_diff','best_of','score','pbp','winner']]
df['tourney_name'] = [s if s==s else 'Davis Cup' for s in df['tourney_name']]
#df = df.drop(['match_num','w_elo','l_elo'],1)
df.head()
```

```python
# DO WE NEED TO FIX THE CODE FOR CURRENT GAME SCORE??
n_samples = sum([len(a)-1 for a in df_pbp3['set_order']])
outcomes = np.zeros(n_samples); counter = 0
set_scores = [0] * n_samples
set_differentials = [0] * n_samples
#game_differentials = [0] * n_samples
elo_differentials = [0] * n_samples
for i in xrange(len(df_pbp3)):
    sets = df_pbp3['pbp'][i].split('.')
    diff = 0; set_diff = 0
    for k in xrange(len(df_pbp3['set_order'][i])-1):
        outcomes[counter] = df_pbp3['winner'][i]
        set_scores[counter] = ''.join([str(a) for a in df_pbp3['set_order'][i][:k+1]])
        score = get_current_game_score(sets[k] + ';')
        diff += score[0]; diff -= score[1]
        set_diff += (2*(df_pbp3['set_order'][i][k]==0) - 1)
        set_differentials[counter] = set_diff
        #game_differentials[counter] = diff
        elo_differentials[counter] = df_pbp3['elo_diff'][i]
        counter += 1
```

```python
# now, 0 means player 0 lost; 1 means player 0 won
df_sets = pd.DataFrame()
df_sets['set_scores'] = set_scores
df_sets['set_diff'] = set_differentials
df_sets['elo_diff'] = elo_differentials
df_sets['outcome'] = [1 - int(a) for a in outcomes]
```
